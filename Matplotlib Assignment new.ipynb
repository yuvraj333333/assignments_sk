Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular plotting library in Python that provides a comprehensive set of tools for creating static, animated, and interactive visualizations. It is widely used for data visualization and is capable of creating various types of plots, charts, and graphs.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

    Line plot: Used to visualize the relationship between two variables, typically with one variable on the x-axis and the other on the y-axis. It shows the trend and continuity of data points.

    Scatter plot: Used to display the relationship between two continuous variables. It represents individual data points as dots on a 2D plane, with one variable on the x-axis and the other on the y-axis.

    Bar plot: Used to compare different categories or groups. It displays the values associated with each category as rectangular bars, making it easy to compare the magnitudes across categories.

    Histogram: Used to visualize the distribution of a continuous variable. It shows the frequency or count of data points falling within specified bins or intervals.

    Pie chart: Used to represent the proportions or percentages of different categories within a dataset. It displays each category as a slice of a circular pie, with the size of the slice proportional to its value.

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a concise summary of the distribution of a dataset. It displays key statistical measures, such as the median, quartiles, and outliers, in a compact and intuitive manner. Box plots are particularly useful for comparing multiple datasets or analyzing the spread and variability of a single dataset.

Here are some key aspects and reasons why box plots are used:

    Visualizing the distribution: A box plot presents a summary of the data distribution, highlighting the central tendency, spread, and skewness. It provides a quick and informative overview of how the data is spread and identifies any potential outliers.

    Identifying central tendency: The box in the plot represents the interquartile range (IQR), which contains the middle 50% of the data. The line inside the box represents the median, which indicates the central tendency of the dataset. Comparing medians between multiple box plots can help understand differences in the central values of different groups.

    Assessing spread and variability: The length of the box indicates the spread of the data, with a longer box representing greater variability. The whiskers, extending from the box, display the range of the dataset excluding outliers. Box plots allow for a quick comparison of the spread and variability between different datasets or groups.

    Detection of outliers: Box plots include "whiskers" that extend from the box to identify potential outliers, which are data points that fall outside the whisker range. Outliers are shown as individual points beyond the whiskers, allowing easy identification of extreme values in the dataset.

    Comparing multiple datasets: Box plots are particularly useful for comparing multiple datasets or groups. By placing multiple box plots side by side, it becomes easy to compare the central tendencies, spreads, and distributions of different groups simultaneously.

    Robust to skewed data: Box plots are robust to skewed data and outliers, making them useful for visualizing datasets that may not follow a normal distribution. They provide a clearer picture of the data distribution, even in the presence of extreme values.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot(box1)
plt.boxplot(box2)

plt.tight_layout()
plt.show()
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(10,9))

plt.boxplot(box1)
plt.boxplot(box2)
```
