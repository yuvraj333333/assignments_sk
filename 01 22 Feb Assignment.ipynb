Q4. What is DataFrame in pandas and how is it different from pandas.series? Explain with an example.

- DataFrame is two-dimensional size,mutable,tabular data structure which can have multiple columns and rows
- Pandas series is one-dimensional array of data & can have only one column of data.

```python
import pandas as pd

# create a Pandas Series with three elements
ser = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

# create a Pandas DataFrame with three columns
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}, index=['x', 'y', 'z'])

# print the Series and the DataFrame
print(ser)
print(df)
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

Both Series and DataFrame are mutable in nature in Pandas, meaning that the values in the object can be changed or updated after it is created. On the other hand, Panel is deprecated in the latest version of Pandas, so it is not relevant to consider its mutability.

Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
# create three series with the same index
s1=pd.Series([1,2,3],index=['a','b','c'])
s2=pd.Series([4,5,6],index=['a','b','c'])
s3=pd.Series([7,8,9],index=['a','b','c'])

# combine the Series into a single DataFrame
df=pd.DataFrame({'S1':s1,'S2':s2,'S3':s3})
print(df)
```
